```python
# Преоразовать существующий код для распознавания последовательности рукописных символов в текст.
# Написать генератор произвольной последовательности рукописных символов.

# ГЕНЕРАЦИЯ НАБОРА ДАННЫХ
# Сгенерировать набор текстовых строк из допустимых символов.
# Каждому набору поставить в соотвествие массив изоражений соответствующих символов.
# Добавить пробелы для разделения слов, символы пробелов не использовать во время классификации.

# КЛАССИФИКАЦИЯ (k ближайших соседей) (пример AIM practice 2)
# Часть символов оставить для обучения классификатора . Остальные  классифицировать
# по аналогии с примером mnist с прошлого занятия. Т.е. преобразовывать изображение в вектор.
# Выбрать оптимальный классификатор в зависимости от k- количество соседей.
# Метрика для измерения точности классификации, и выбора оптимального классификатора относительно параметра k - accuracy.

# РАСПОЗНАВАНИЕ
# заменить inf_net на последовательное применение следующих шагов:
# 1. генерация изображения с последовательностью символов
# 2. Выделение объектов оставляете как есть.
# 3. Примернить k-nn для решения задачи классификации. 
# 4. Вывести результат
```

```python
#КЛАССИФИКАЦИЯ (k-NN)

import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

DATA_DIR = r'C:\Users\админ\labb9\data'
IMG_SIZE = 32
MAX_PER_CLASS = 200

def imread_unicode(path):
    data = np.fromfile(path, dtype=np.uint8)
    return cv2.imdecode(data, cv2.IMREAD_COLOR)

def preprocess_symbol(img_bgr, out_size=IMG_SIZE):
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    inv = cv2.bitwise_not(gray)
    _, bw = cv2.threshold(inv, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    ys, xs = np.where(bw > 0)
    if len(xs) == 0:
        return cv2.resize(bw, (out_size, out_size))

    crop = bw[ys.min():ys.max()+1, xs.min():xs.max()+1]
    s = max(crop.shape)
    pad = ((s - crop.shape[0])//2, (s - crop.shape[1])//2)
    square = cv2.copyMakeBorder(
        crop, pad[0], s-crop.shape[0]-pad[0],
        pad[1], s-crop.shape[1]-pad[1],
        cv2.BORDER_CONSTANT, value=0
    )
    return cv2.resize(square, (out_size, out_size))

files_by_char = {}
for folder in os.listdir(DATA_DIR):
    path = os.path.join(DATA_DIR, folder)
    if not os.path.isdir(path):
        continue
    ch = '*' if folder == 'times' else folder
    files = [
        os.path.join(path, f)
        for f in os.listdir(path)
        if f.lower().endswith('.jpg')
    ]
    files_by_char[ch] = files[:MAX_PER_CLASS]

allowed_chars = sorted(files_by_char.keys())
print("Классы:", allowed_chars)

train_files, test_files = {}, {}
for ch, files in files_by_char.items():
    tr, te = train_test_split(files, test_size=0.3, random_state=42)
    train_files[ch] = tr
    test_files[ch] = te

def build_Xy(files_dict):
    X, y = [], []
    for ch, files in files_dict.items():
        for fp in files:
            img = imread_unicode(fp)
            if img is None:
                continue
            bw = preprocess_symbol(img)
            X.append((bw > 0).reshape(-1).astype(np.float32))
            y.append(ch)
    return np.array(X), np.array(y)

X_train, y_train = build_Xy(train_files)
X_test, y_test = build_Xy(test_files)

# Обучение
knn_models = {}
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)
for k in [1, 3, 5]:
    model = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    model.fit(X_train, y_train_enc)
    acc = accuracy_score(y_test_enc, model.predict(X_test))
    knn_models[k] = model
    print(f"k={k} accuracy={acc:.4f}")

best_k = max(knn_models, key=lambda k: accuracy_score(y_test, knn_models[k].predict(X_test)))
best_knn = knn_models[best_k]
print(f"\nЛучшая модель: k={best_k}")
```

```python
# ГЕНЕРАЦИЯ

import numpy as np
import matplotlib.pyplot as plt
from random import choice, random
import cv2

def generate_text_from_train(min_len=6, max_len=10, space_prob=0.25):
    L = np.random.randint(min_len, max_len+1)
    text = []
    picked = []

    for i in range(L):
        if i > 0 and random() < space_prob:
            text.append(' ')
            picked.append((' ', None))

        ch = choice(allowed_chars)
        fp = choice(train_files[ch])
        img = imread_unicode(fp)
        if img is None:
            continue

        bw = preprocess_symbol(img, IMG_SIZE)  # 32x32 бинарное (0/255)
        text.append(ch)
        picked.append((ch, bw))

    return ''.join(text), picked

def render_line_from_picked(picked, gap=6, space_gap=30):
    tiles = []
    for ch, bw in picked:
        if ch == ' ':
            tiles.append(np.zeros((IMG_SIZE, space_gap), dtype=np.uint8))
        else:
            tiles.append(bw)
            tiles.append(np.zeros((IMG_SIZE, gap), dtype=np.uint8))
    return np.hstack(tiles)

# Генерация
text_gt, picked = generate_text_from_train()
img_line = render_line_from_picked(picked)

print("Сгенерированная строка (GT):", repr(text_gt))

picked_only = [(ch, bw) for ch, bw in picked if ch != ' ']
plt.figure(figsize=(12,2))
m = min(10, len(picked_only))
for i in range(m):
    ch, bw = picked_only[i]
    plt.subplot(1, m, i+1)
    plt.imshow(bw, cmap='gray')
    plt.title(ch)
    plt.axis('off')
plt.suptitle("Выбранные изображения символов из train (первые 10)")
plt.tight_layout()
plt.show()

# Картинка сгенерированной строки
plt.figure(figsize=(12,2))
plt.imshow(img_line, cmap='gray')
plt.title("Сгенерированное изображение строки")
plt.axis('off')
plt.show()

# Разделение
print("Количество символов (без пробелов):", len(picked_only))
plt.figure(figsize=(12,2))
m = min(10, len(picked_only))
for i in range(m):
    ch, bw = picked_only[i]
    plt.subplot(1, m, i+1)
    plt.imshow(bw, cmap='gray')
    plt.title(f"Симв.{i+1}")
    plt.axis('off')
plt.suptitle("Разделённые символы (по генерации) — первые 10")
plt.tight_layout()
plt.show()
bw = img_line.copy().astype(np.uint8)
contours, _ = cv2.findContours(bw, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

boxes = []
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if w*h > 150:
        boxes.append((x,y,w,h))

boxes.sort(key=lambda t: t[0])

plt.figure(figsize=(12,2))
for i,(x,y,w,h) in enumerate(boxes[:10]):
    crop = bw[y:y+h, x:x+w]
    plt.subplot(1,len(boxes[:10]),i+1)
    plt.imshow(crop, cmap='gray')
    plt.title(f"Obj{i+1}")
    plt.axis('off')
plt.suptitle("Выделение объектов (контуры)")
plt.tight_layout()
plt.show()
```

```python
# РАСПОЗНАВАНИЕ + ОЦЕНКА МОДЕЛЕЙ

def recognize_picked_with_model(picked, model):
    out = []
    for ch, bw in picked:
        if ch == ' ':
            out.append(' ')
            continue
        vec = ((bw > 0).reshape(1, -1).astype(np.float32))
        pred_idx = model.predict(vec)[0]               
        pred_ch = le.inverse_transform([pred_idx])[0]
        out.append(pred_ch)
    return ''.join(out)

def acc_chars(gt, pred):
    gt_n = gt.replace(' ', '')
    pr_n = pred.replace(' ', '')
    L = min(len(gt_n), len(pr_n))
    correct = sum(1 for i in range(L) if gt_n[i] == pr_n[i])
    return correct / len(gt_n) if len(gt_n) else 0.0

print("GT:", repr(text_gt))

results = []
for k, model in knn_models.items():
    pred = recognize_picked_with_model(picked, model)
    a = acc_chars(text_gt, pred)
    results.append((k, a, pred))
    print(f"\n[k={k}] PRED={repr(pred)}")
    print(f"Точность по символам (без пробелов): {a:.4f} | exact={pred == text_gt}")

best = max(results, key=lambda x: x[1])
print(f"\nЛучшая модель по распознаванию строки: k={best[0]} (acc_chars={best[1]:.4f})")
```
