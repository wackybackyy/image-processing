```python
# 1. Загрузите изображение в оттенках серого sar_1_gray.jpg.
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("sar_1_gray.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
# 2. постройте гистограмму
import numpy as np

plt.hist(image.ravel(), bins=256, range=(0, 256))
plt.show()
```

```python
# 3. реализуйте алгоритм гамма коррекции с параметром гамма <1, >1.
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(img, gamma):
    return np.uint8(255 * (img / 255) ** gamma)

gamma1, gamma2 = 0.5, 2.0
img_light = gamma_correction(image, gamma1)
img_dark = gamma_correction(image, gamma2)

titles = ['Оригинал', 'Гамма=0.5', 'Гамма=2.0']
images = [image, img_light, img_dark]

plt.figure(figsize=(10, 4))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])

plt.tight_layout()
plt.show()
```

```python
# 4. Сравните исходное изображение, скорректированное при помощи гамма-фильтра. MSE, SSIM.
from skimage.metrics import mean_squared_error, structural_similarity
import matplotlib.pyplot as plt

print("MSE (Гамма=0.5):", mean_squared_error(image, img_light))
print("MSE (Гамма=2.0):", mean_squared_error(image, img_dark))

print("SSIM (Гамма=0.5):", structural_similarity(image, img_light, channel_axis=None, win_size=3))
print("SSIM (Гамма=2.0):", structural_similarity(image, img_dark, channel_axis=None, win_size=3)) 
```

```python
# 5. реализуйте алгоритм статистической цветокоррекции на основе статистики eq_gray.
import numpy as np
import matplotlib.pyplot as plt

eq_gray = cv2.imread("sar_1_gray.jpg", cv2.IMREAD_GRAYSCALE)

def statistical_color_correction(src, ref):
    mean_src, std_src = np.mean(src), np.std(src)
    mean_ref, std_ref = np.mean(ref), np.std(ref)
    
    corrected = (src - mean_src) * (std_ref / std_src) + mean_ref
    corrected = np.clip(corrected, 0, 255)
    
    return corrected.astype(np.uint8)

corrected_img = statistical_color_correction(image, eq_gray)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.imshow(image, cmap='gray'); plt.title("Оригинал")
plt.subplot(1,3,2); plt.imshow(eq_gray, cmap='gray'); plt.title("eq_gray")
plt.subplot(1,3,3); plt.imshow(corrected_img, cmap='gray'); plt.title("После коррекции")
plt.tight_layout()
plt.show()
```

```python
# 6. Протестируйте работу алгоритмов пороговой фильтрации с различными параметрами.
import cv2
import matplotlib.pyplot as plt

# Ручные значения порога
threshold_values = [50, 100, 150, 200]
binary_images = []

plt.figure(figsize=(15, 6))

#1. Пороговая фильтрация
for i, T in enumerate(threshold_values):
    _, thresh = cv2.threshold(image, T, 255, cv2.THRESH_BINARY)
    binary_images.append(thresh)
    plt.subplot(2, len(threshold_values) + 2, i + 1)
    plt.imshow(thresh, cmap='gray')
    plt.title(f"Порог={T}")
    plt.axis('off')

#2.Метод Отсу (автоматический)
_, otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.subplot(2, len(threshold_values) + 2, len(threshold_values) + 1)
plt.imshow(otsu, cmap='gray')
plt.title(f"Отсу (T={_:.0f})")
plt.axis('off')

#3.Адаптивная пороговая фильтрация
adaptive = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY, 11, 2)
plt.subplot(2, len(threshold_values) + 2, len(threshold_values) + 2)
plt.imshow(adaptive, cmap='gray')
plt.title("Адаптивная")
plt.axis('off')

plt.tight_layout()
plt.show()

```
